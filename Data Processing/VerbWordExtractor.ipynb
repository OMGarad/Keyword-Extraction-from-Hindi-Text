```python
file1= open(r"idxverb_txt.txt", "r", encoding='utf-8')
i= file1.readlines()
```

```python
print(i)
```

```python
l=[]
for line in i:
    index=line.find(' ')
    if index!=-1:
        shortenedLines= line[:index]
        shortenedSeparatedLines= shortenedLines + "\n"
        l.append(shortenedSeparatedLines)
        print(shortenedSeparatedLines)
```
